# 1. Create 'price_label' and 'busiest_day' columns

## Creating 'price_label' Column and Ranking products based on their prices

### Using If-Statement and LOC() Function for the following Price Ranges

```python
# Low range product where prices are below 5
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] <= 5, 'price_label'] = 'Low range products'
# Mid range products where prices are between 15 and 5
df_ords_prods_merge.loc[(df_ords_prods_merge['prices'] <= 15) & (df_ords_prods_merge['prices'] > 5), 'price_label'] = 'Mid range products'
## High range products where price is more than 15
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] > 15, 'price_label'] = 'High range products'
```

## Creating 'busiest_day' column and ranking order_day_of_week depending how busy they were

### Using For Loops

```python
# Frequency Count for order_day_of_week
df_ords_prods_merge['order_day_of_week'].value_counts(dropna = False)
```

```python
# Using for loops to create a new list

result = []

for x in df_ords_prods_merge['order_day_of_week']:
    if x == 0:
        result.append('Busiest day')
    elif x == 4:
        result.append('Least busy')
    else:
        result.append('Regulary busy')
```

```python
# Verifying the list

result
```

```python
## Comibining with the ords_prods_merge
df_ords_prods_merge['busiest_day'] = result
# Frequency Count
df_ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

# 2. Updating the 'busiest_day' to 'busiest_days' and Ranking top 2 Busiest days and Top 2 Slowest Days

## Changing the column name to 'busiest_days'

```python
# Changing the column name
df_ords_prods_merge.rename(columns = {'busiest_day':'busiest_days'}, inplace = True)
```

```python
# Verifying if the column name changed
df_ords_prods_merge.head()
```

## Creating a new category for busiest days, slowest days and regularly busy

```python
# Frequency Count for order_day_of_week
df_ords_prods_merge['order_day_of_week'].value_counts(dropna = False)
```

```python
# Using For Loops for creating a new list where 0 OR 1 are busiest days, 3 OR 4 are slowest dayys and else as regularly busy

result_new = []

for y in df_ords_prods_merge['order_day_of_week']:
    if y == 0 or y == 1:
        result_new.append('Busiest days')
    elif y == 4 or y == 3:
        result_new.append('Slowest days')
    else:
        result_new.append('Regularly busy')
```

```python
# # Comibining with the ords_prods_merge
df_ords_prods_merge['busiest_days'] = result_new
```

# 4. Create a column 'busiest_period_of_day' and Label 'order_hour_of_day' As Most orders, Average orders, Fewest orders.

## Frequency Count of 'order_hour_of_day'

```python
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

## Using For Loops for creating a new list


```python
# Value_counts Between 2761760 and 2000000 = Most orders
# Value_counts Between 2000000 and 400000 = Average orders
# Value_counts Below 300000 = Fewest orders

result_busiest_time_of_day = []

for z in df_ords_prods_merge['order_hour_of_day']:
    if z in [10,11,14,15,13,12,16,9,17 ]:
        result_busiest_time_of_day.append('Most orders')
    elif z in [6,0,1,5,2,4,3]:
        result_busiest_time_of_day.append('Fewest orders')
    else:
        result_busiest_time_of_day.append('Average orders')

```

```python
# Verifying the new list
result_busiest_time_of_day
```

```python
# Comibining with the df_ords_prods_merge
df_ords_prods_merge['busiest_period_of_day'] = result_busiest_time_of_day
```

```python
# Verifying the dataframe with the added column

df_ords_prods_merge.head()
```
