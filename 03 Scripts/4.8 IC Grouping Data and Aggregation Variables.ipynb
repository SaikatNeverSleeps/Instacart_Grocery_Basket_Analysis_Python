# 4. Creating loyalty flag for existing customers

## 4.1. Creating 'max_order' column

```python
# Creating 'max_order' column on 'order_number' by using transform() function grouped by 'user_id'
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby (['user_id']) ['order_number'].transform(np.max)
```

```python
# Verifying column creation
df_ords_prods_merge[['max_order']] .head(20)
```

```python
# Frequency Count for 'max_order'
df_ords_prods_merge['max_order'].value_counts(dropna = False)
```

## 4.2. Creating 'loyalty_flag' column by deriving Columns with loc()

```python
# If the maximum orders the user has made is over 40, then the customer will be labeled a “Loyal customer.”
# If the maximum orders the user has made is over 10 but less than or equal to 40, then the customer will be labeled a “Regular customer."
# If the maximum orders the user has made is less than or equal to 10, then the customer will be labeled a “New customer.”

df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Frequency Coount of loyalty Flag
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Verifying column 
df_ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

# 5. Basic statistics of the product prices for each loyalty category

```python
# Max price of 99999.0 seems odd and seems like an outlier
# Finding outliers for 'price' with the condition that no price should be more than 100
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] > 100]
```

```python
# There are 5127 counts where the products are more than 100, of which we can see 149000 and 99999.0
# Marking the outliers as NaNs
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] > 100, 'prices'] = np.nan
```

```python
# Verifying changes
df_ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean', 'min', 'max']})
```

```python
# Finding the value_counts() of 'price_label' grouped by 'loyalty_flag'
df_ords_prods_merge.groupby('loyalty_flag')['price_label'].value_counts().unstack()
```

### Based on the first table, loyal customers tends buy cheaper products while new customer tends to buy pricier products. This is also evident by the second table where loyal customers tends to buy more Low Range products and Mid range products. It is quiet evident that all customers tend to buy Mid range range products.

# 6. Creating spending flag for existing customers

## 6.1. Creating 'average_product_price' column for each customer

```python
# Creating 'average_price' column on 'prices' by using transform() function grouped by 'user_id'
df_ords_prods_merge['average_product_price'] = df_ords_prods_merge.groupby (['user_id']) ['prices'].transform(np.max)
```

```python
# Verifying column creation
df_ords_prods_merge[['user_id', 'average_product_price']].head(20)
```

## 6.2. Creating 'spending_flag' column by deriving Columns with loc()

```python
# If 'average_produc_price' is less than 10 = 'Low spender'
df_ords_prods_merge.loc[df_ords_prods_merge['average_product_price'] < 10, 'spending_flag'] = 'Low spender'
# If 'average_produc_price' is more than OR equal to 10  = 'High spender'
df_ords_prods_merge.loc[(df_ords_prods_merge['average_product_price'] >= 10), 'spending_flag'] = 'High spender'
```

```python
# Frequency Coount of loyalty Flag
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Verifying column 
df_ords_prods_merge[['user_id', 'average_product_price', 'spending_flag']].head(60)
```

# 7. Creating order frequency flag for existing customers

## 7.1. Creating 'median_reorder' column

```python
# Creating 'median_reorder' column on 'days_since_prior_order' by using transform() function grouped by 'user_id'
df_ords_prods_merge['median_reorder'] = df_ords_prods_merge.groupby (['user_id']) ['days_since_prior_order'].transform(np.median)
```

```python
# Verifying column creation
df_ords_prods_merge[['user_id', 'median_reorder']].head(20)
```

## 6.2. Creating 'order_frequency_flag' column by deriving Columns with loc()

```python
# If 'median_reorder' is higher than 20 = 'Non-frequent customer'
df_ords_prods_merge.loc[df_ords_prods_merge['median_reorder'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
# If 'median_reorder' is higher than 10 AND lower than OR equal to 20 = 'Regular customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['median_reorder'] > 10) & (df_ords_prods_merge['median_reorder'] <= 20), 'order_frequency_flag'] = 'Regular customer'
# If 'median_reorder' is lower OR equal to 10 = 'Frequent customer'
df_ords_prods_merge.loc[df_ords_prods_merge['median_reorder'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# Frequency Coount of order_frequency_flag
df_ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
# Verifying column 
df_ords_prods_merge[['user_id', 'median_reorder', 'order_frequency_flag']].head(60)
```
