# Defining Variable and if_else statements

```python
def price_lable(row):
    if row ['prices'] <= 5:
        return 'Low-range products'
    elif row ['prices'] > 5 and row ['prices'] <= 15:
        return 'Mid-range products'
    elif row ['prices'] > 15:
        return 'High range'
    else:
        return 'Not enough data'
```

## Aplying the function

```python
df['price_range'] = df.apply(price_lable, axis=1)
```

```python
# Frequency Count based on the price range
df['price_range'].value_counts(dropna = False)
```

## Using the loc() function 


```python
# High range products
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High range products'
```

```python
# Mid range products
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid range products'
```

```python
# Low range products
df.loc[df['prices'] < 5, 'price_range_loc'] = 'Low range products'
```

```python
# Frequency Count
df['price_range_loc'].value_counts(dropna = False)
```
